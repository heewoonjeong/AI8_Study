```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = np.load('fruits_300.npy')
```

```python
print(fruits.shape) 
```

```python
print(fruits[0, 0, :])
```

```python
plt.imshow(fruits[0], cmap= 'gray') # 컴퓨터랑 사람이 인식하는게 다르다 
plt.show() # 배경을 검은색으로 바꿈 사과가 알고싶은거라서 
```

```python
plt.imshow(fruits[0],cmap = 'gray_r') # 보기좋게 만듬 
plt.show()
```

```python
fig, axs = plt.subplots (1,2) # 한열에 2컬럼을 출력 
axs[0].imshow(fruits[100], cmap = 'gray_r')
axs[1].imshow(fruits[200], cmap = 'gray_r')
plt.show()
```

```python
apple =fruits[0:100].reshape( -1, 100*100) # 1차원으로 배치해서 인식하기 쉽게 변경 
pineapple = fruits[100:200].reshape( -1 ,100*100) # 순서대로하기위해 슬라이싱연산자 사용 
banana = fruits[200:300].reshape(-1, 100*100)
print(apple.shape) 
```

```python
print(apple.mean(axis=1))  #애플의 평균 계산 
```

```python
plt.hist(np.mean(apple,axis=1), alpha = 0.7)
plt.hist(np.mean(pineapple, axis =1), alpha=0.7)
plt.hist(np.mean(banana, axis =1), alpha=0.7) #알파>그래프  투명도
plt.legend(['apple','pineapple','banana'])
plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))
axs[0].bar(range(10000),np.mean(apple,axis =0))
axs[1].bar(range(10000),np.mean(pineapple,axis=0))
axs[2].bar(range(10000),np.mean(banana,axis=0))
plt.show() 
```

```python
apple_mean = np.mean(apple, axis=0).reshape(100, 100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100,100)
banana_mean = np.mean(banana, axis=0).reshape(100,100)
fig, axs = plt.subplots(1,3, figsize=(20,5)) # 평균값으로 이미지 구현 
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pineapple_mean, cmap = 'gray_r')
axs[2].imshow(banana_mean, cmap = 'gray_r')
plt.show()
```

```python
abs_diff = np.abs(fruits - apple_mean) # abs 절댓값을 계산하는 함수 
abs_mean = np.mean(abs_diff, axis = (1,2)) #모든 절댓값과 사과의 절댓값의 오차를 사용 
print(abs_mean.shape) # 모든 샘플에서 apple mean을 뺀 절댓값의 평ㄱㄴ을 계산 
```

```python
apple_index = np.argsort(abs_mean)[:100] # 오차가 가장작은 것들을 그래프로 그림 
fig, axs = plt.subplots(10,10, figsize=(10,10))
for i in range(10) : 
  for j in range(10) : 
    axs[i, j].imshow(fruits[apple_index[i*10 + j]], cmap= 'gray_r')
    axs[i, j].axis('on')
plt.show()
```

```python
abs_diff = np.abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff,axis=(1,2))

banana_index = np.argsort(abs_mean)[:100]
fig, axs = plt.subplots(10,10, figsize=(10,10))
for i in range(10): 
  for j in range(10) : 
    axs[i,j].imshow(fruits[banana_index[i*10 + j]], cmap='gray_r')
    axs[i,j].axis('off')
plt.show()
```
